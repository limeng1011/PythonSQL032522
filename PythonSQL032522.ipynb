```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('sakila.db')
```

```python
sql_query = """SELECT * from customer;"""
df = pd.read_sql(sql_query, con)
```

```python
df.head()
```

```python
sql_query2 = """SELECT first_name, last_name from customer;"""
df2 = pd.read_sql(sql_query2, con)
```

```python
df2.head()
```

```python
def sql_to_df(query):
  return pd.read_sql(query,con)
```

```python
sql_to_df(sql_query2).head()
```

```python
# Select distinct country_ids rom the city table
sql_query3 = """SELECT DISTINCT(country_id) FROM city;"""
sql_to_df(sql_query3).head()
```

```python
# Select all customer info from the 1st store.
sql_query4 = """SELECT * FROM customer WHERE store_id = 1"""
sql_to_df(sql_query4).head()
```

```python
sql_query5 = """SELECT * FROM customer WHERE first_name = 'MARY';"""
sql_to_df(sql_query5).head()
```

```python
# Select all films from 2006 that are rated R
sql_query6 = """SELECT * FROM film WHERE release_year = 2006 AND rating = 'R';"""
sql_to_df(sql_query6).head()
```

```python
# Select all films from R & PG.
sql_query7 = """SELECT * from film WHERE rating = 'R' OR rating = 'PG';"""
sql_to_df(sql_query7).head()
```

```python
# Count the number of customers
sql_query8 = """SELECT customer_id,COUNT(customer_id) FROM customer;"""
sql_to_df(sql_query8)
```

```python
# The % wildcard
# Select any customers whose name start with an M
sql_query9 = """SELECT * FROM customer WHERE first_name LIKE 'M%';"""
sql_to_df(sql_query9).head()
```

```python
# The _ wildcard
# Select any customers whole last name ends with ing
sql_query10 = """SELECT * FROM customer WHERE last_name LIKE '_ING';"""
sql_to_df(sql_query10).head()
```

```python
"""
The [character_list] wildcard
In MySQL you would use: 
WHERE value LIKE '[charlist]%'
IN SQLite you use:
WHERE value GLOB '[charlist]*'
"""
# Select any customers whose first name begins with an A or a B
sql_query11 = """SELECT * FROM customer WHERE first_name GLOB '[AB]*';"""
sql_to_df(sql_query11).head()
```

```python
# Select all customers and order results by last name
sql_query12 = """SELECT * FROM customer ORDER BY last_name;"""
sql_to_df(sql_query12).head()
```

```python
sql_query13 = """SELECT * FROM customer ORDER BY last_name DESC;"""
sql_to_df(sql_query13).head()
```

```python
# Count the number of customers per store
sql_query14 = """SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id;"""
sql_to_df(sql_query14)
```
